```python
import networkx as nx
import matplotlib.pyplot as plt
```

Networkx is a library for node creation

```python
g=nx.Graph()
# g is an empty graph with no nodes and edges
```

```python
# to add nodes and edges
# to create a node we need to write add_node and for edge we will write add_edge
g.add_node(2)
g.add_node(5)
g.add_edge(2,5)
g.add_edge(4,1)
#to add a list of edges we can also write 
# # this in it in this way
g.add_edges_from([(2,5),(1,3)])
```

```python
# print (nx.info(g))
nx.draw(g)
plt.show()
```

```python
# to add nodes 
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
#to add a list of edges we can also write 
# this in it in this way
```

```python
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(3,4)
g.add_edge(4,1)
```

```python
# print (nx.info(g))
nx.draw(g)
c=nx.Graph()
'''The spring_layout function is particularly
useful for visualizing large and complex networks where the relationships between nodes need to be clearly understood and represented'''
pos=nx.spring_layout(g)
nx.draw(c, pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, font_color='black', font_weight='bold', edge_color='gray', width=2.0)
plt.title('Rectangle')

plt.show()
```

```python
edges_list = [(1, 2), (2, 3), (3, 4), (4, 1)]  # Define edges to form a rectangle
g.add_edges_from(edges_list)  # Add edges to the graph

# Draw the graph
pos = nx.circular_layout(g)  # Compute node positions
nx.draw(g, pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, font_color='black', font_weight='bold', edge_color='gray', width=2.0)

# Add a title
plt.title('Rectangle-like Structure')

# Display the graph
plt.show()
```
